```python
# Import libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# =============================
# Load Dataset
# =============================
file_path = r"C:\Users\Dell\Documents\Modified_Crime_Dataset.csv"
data = pd.read_csv(file_path)

# =============================
# Preprocessing: Binary Encoding
# =============================
# Select relevant columns (e.g., crime categories)
crime_columns = data.columns[3:]  # Adjust column range based on your dataset structure

# Binarize data: Convert non-zero values to 1 (True) and zero to 0 (False)
binary_data = (data[crime_columns] > 0).astype(bool)

# Debug: Print sample of binary data
print("Binary Data Sample:")
print(binary_data.head())

# =============================
# Dimensionality Reduction
# =============================
# Reduce columns to top features with high variance (optional)
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.01)  # Adjust threshold if needed
binary_data_reduced = selector.fit_transform(binary_data)

# Debug: Print reduced dataset shape
print("Reduced Binary Data Shape:", binary_data_reduced.shape)

# =============================
# Apply Apriori Algorithm
# =============================
# Set a higher min_support to reduce memory usage
frequent_itemsets = apriori(pd.DataFrame(binary_data_reduced), 
                            min_support=0.1,  # Adjust as needed (higher reduces memory usage)
                            use_colnames=True, 
                            low_memory=True)

# Debug: Print frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)

# =============================
# Generate Association Rules
# =============================
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Debug: Print association rules
print("Association Rules:")
print(rules)

# =============================
# Save Results
# =============================
# Save frequent itemsets and rules to CSV files
frequent_itemsets.to_csv(r"C:\Users\Dell\Documents\frequent_itemsets.csv", index=False)
rules.to_csv(r"C:\Users\Dell\Documents\association_rules.csv", index=False)

print("Frequent itemsets and rules saved to CSV files.")
```
